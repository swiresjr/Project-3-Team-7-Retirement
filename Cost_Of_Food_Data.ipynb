```python
#Import dependencies

import pandas as pd
from pathlib import Path
import numpy as np
```

```python
#Import csv files and create dataframes

cost_of_living_path = "Resources/cost_of_living_us.csv"

cost_stats = pd.read_csv(cost_of_living_path)
cost_stats
```

```python
#limit family_member_count to 2p0c

cost_stats_2p0c = cost_stats[cost_stats['family_member_count'] == '2p0c']
cost_stats_2p0c
```

```python
food_cost_stats = cost_stats_2p0c.drop(columns=[
    'case_id', 'isMetro', 'areaname', 'county', 'family_member_count', 
    'housing_cost', 'transportation_cost', 'healthcare_cost', 
    'other_necessities_cost', 'childcare_cost', 'taxes', 'total_cost'
])

food_cost_stats
```

```python
#Rename columns

food_cost_stats_renamed = food_cost_stats.rename(columns={
    'state': 'State',
    'food_cost': 'Food Cost',
    'median_family_income': 'Median Family Income'
})

food_cost_stats_renamed
```

```python
# Group by 'State' and calculate the averages for each state
food_state_averages = food_cost_stats_renamed.groupby('State').agg({
    'Food Cost': 'mean',
    'Median Family Income': 'mean'
}).reset_index()


food_state_averages.rename(columns={
    'Food Cost': 'Average Food Cost Per Year',
    'Median Family Income': 'Average Median Family Income Per Year'
}, inplace=True)


food_state_averages
```

```python
#Sort ascending 
food_state_averages_sorted = food_state_averages.sort_values(by='Average Food Cost Per Year').reset_index(drop=True)

food_state_averages_sorted
```

```python
#convert to $ format

# Verify all columns are numeric
food_state_averages_sorted['Average Food Cost Per Year'] = pd.to_numeric(
    food_state_averages_sorted['Average Food Cost Per Year'], errors='coerce'
)
food_state_averages_sorted['Average Median Family Income Per Year'] = pd.to_numeric(
    food_state_averages_sorted['Average Median Family Income Per Year'], errors='coerce'
)

# Apply dollar formatting to valid numeric values
food_state_averages_sorted['Average Food Cost Per Year'] = food_state_averages_sorted['Average Food Cost Per Year'].apply(
    lambda x: f"${x:,.2f}" if pd.notnull(x) else "N/A"
)
food_state_averages_sorted['Average Median Family Income Per Year'] = food_state_averages_sorted['Average Median Family Income Per Year'].apply(
    lambda x: f"${x:,.2f}" if pd.notnull(x) else "N/A"
)

food_state_averages_sorted
```

```python
#export to csv

output_path = "food_state_averages_sorted.csv"
food_state_averages_sorted.to_csv(output_path, index=False)

output_path
```
