```python
#Import dependencies

import pandas as pd
from pathlib import Path
import numpy as np
```

```python
#Import csv files and create dataframes

cost_of_living_path = "Resources/cost_of_living_us.csv"

cost_stats = pd.read_csv(cost_of_living_path)
cost_stats
```

```python
#limit family_member_count to 2p0c

cost_stats_2p0c = cost_stats[cost_stats['family_member_count'] == '2p0c']
cost_stats_2p0c
```

```python
housing_cost_stats = cost_stats_2p0c.drop(columns=[
    'case_id', 'isMetro', 'areaname', 'county', 'family_member_count', 
    'food_cost', 'transportation_cost', 'healthcare_cost', 
    'other_necessities_cost', 'childcare_cost', 'taxes', 'total_cost'
])

housing_cost_stats
```

```python
#Rename columns

housing_cost_stats_renamed = housing_cost_stats.rename(columns={
    'state': 'State',
    'housing_cost': 'Housing Cost',
    'median_family_income': 'Median Family Income'
})

housing_cost_stats_renamed
```

```python
# Group by 'State' and calculate the averages for each state
housing_state_averages = housing_cost_stats_renamed.groupby('State').agg({
    'Housing Cost': 'mean',
    'Median Family Income': 'mean'
}).reset_index()


housing_state_averages.rename(columns={
    'Housing Cost': 'Average Housing Cost Per Year',
    'Median Family Income': 'Average Median Family Income Per Year'
}, inplace=True)


housing_state_averages
```

```python
#Sort ascending 
housing_state_averages_sorted = housing_state_averages.sort_values(by='Average Housing Cost Per Year').reset_index(drop=True)

housing_state_averages_sorted
```

```python
#convert to $ format

housing_state_averages_sorted['Average Housing Cost Per Year'] = housing_state_averages_sorted['Average Housing Cost Per Year'].apply(
    lambda x: f"${x:,.2f}" if pd.notnull(x) else "N/A"
)
housing_state_averages_sorted['Average Median Family Income Per Year'] = housing_state_averages_sorted['Average Median Family Income Per Year'].apply(
    lambda x: f"${x:,.2f}" if pd.notnull(x) else "N/A"
)

housing_state_averages_sorted
```

```python
#export to csv

output_path = "housing_state_averages_sorted.csv"
housing_state_averages_sorted.to_csv(output_path, index=False)

output_path
```
