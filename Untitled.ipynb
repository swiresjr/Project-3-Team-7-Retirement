```python
import pandas as pd
```

```python

# Read the CSV file into a Pandas DataFrame
healthcare_data = pd.read_csv('/Users/phunghuynh/Desktop/Project-3-Team-7-Retirement/Healthcare_cost_per_state.csv')

# Display the first few rows of the DataFrame
print(healthcare_data.head())
```

```python

# Assign "State" as the index
healthcare_data.set_index('State', inplace=True)
healthcare_data.head()
```

```python
# Sort the DataFrame by "Average Annual Premium"
healthcare_data_sorted = healthcare_data.sort_values(by='Average Annual Premium')

# Update the "Rank" column based on the new order (lowest value gets rank 1)
healthcare_data_sorted['Rank'] = range(1, len(healthcare_data_sorted) + 1)

# Reset the index to keep the "State" as the index
healthcare_data_sorted.reset_index(drop=False, inplace=True)

# Set "State" back as the index
healthcare_data_sorted.set_index('State', inplace=True)
healthcare_data_sorted.rename(columns={'Rank': 'Rank_healthcare'}, inplace=True)
healthcare_data_sorted.head()
```

```python
# Read the CSV file into a Pandas DataFrame
food_data = pd.read_csv('/Users/phunghuynh/Desktop/Project-3-Team-7-Retirement/food_state_averages_sorted.csv')
    
# Display the first few rows of the DataFrame
food_data.head()
```

```python
# Remove the column "Average Median Family Income Per Year"
food_data.drop(columns=['Average Median Family Income Per Year'], inplace=True)

# Rename the columns
food_data.columns = ['State', 'Average Annual Food Cost', 'Rank_food']



# Assign "State" as the index
food_data.set_index('State', inplace=True)

# Reorder the columns to put "Rank" before "Average Annual Food Cost"
food_data = food_data[['Rank_food', 'Average Annual Food Cost']]
food_data.head()
```

```python
# Read the CSV file into a Pandas DataFrame
housing_data = pd.read_csv('/Users/phunghuynh/Desktop/Project-3-Team-7-Retirement/housing_state_averages_sorted.csv')
    
# Display the first few rows of the DataFrame
housing_data.head()
```

```python
# Remove the column "Average Median Family Income Per Year"
housing_data.drop(columns=['Average Median Family Income Per Year'], inplace=True)

# Rename the columns
housing_data.columns = ['State', 'Average Annual Housing Cost', 'Rank_housing']



# Assign "State" as the index
housing_data.set_index('State', inplace=True)

# Reorder the columns to put "Rank" before "Average Annual Food Cost"
housing_data = housing_data[['Rank_housing', 'Average Annual Housing Cost']]
housing_data.head()
```

```python
# Read the CSV file into a Pandas DataFrame
crime_data = pd.read_csv('/Users/phunghuynh/Desktop/Project-3-Team-7-Retirement/Crime_stats_2016.csv')
    
# Display the first few rows of the DataFrame
crime_data.head()
```

```python
# Assign “State” column as index
crime_data.set_index("State", inplace=True)
# Rename “Ranking” to “Rank”
crime_data.rename(columns={'Ranking': 'Rank_crime'}, inplace=True)

# Change the “Rank” column value based on the value of “Total Crimes” column
crime_data['Rank_crime'] = crime_data['Total Crimes'].rank(method='min').astype(int)


crime_data.head()
```

```python
# Merge the first two DataFrames on "State"
merged_data = pd.merge(healthcare_data_sorted, food_data, on='State')

# Merge the result with the housing_data DataFrame
merged_data = pd.merge(merged_data, housing_data, on='State')

# Finally, merge the result with the crime_data DataFrame
merged_data = pd.merge(merged_data, crime_data, on='State')

# Display the merged DataFrame
merged_data.head()

```

```python
# Create the "total score" column by summing the rank columns
merged_data['total_score'] = (merged_data['Rank_healthcare'] + 
                               merged_data['Rank_food'] + 
                               merged_data['Rank_housing'] + 
                               merged_data['Rank_crime'])
merged_data.head()
```

```python
# Create the "final rank" column based on the "total_score" column
merged_data['final_rank'] = merged_data['total_score'].rank(method='min').astype(int)

# Sort the DataFrame by "total_score" in ascending order
merged_data.sort_values(by='total_score', inplace=True)

# Export the DataFrame to a CSV file named "master_data.csv"
merged_data.to_csv("master_data.csv", index=True)

merged_data
```
